```r
dataset <- read.csv('D:/machine_learning/Startups.csv')
```

```r
dataset
```

```r
dataset$State = factor(dataset$State,
                      levels = c('New York','Florida','California'),
                      labels = c(1,2,3))
```

```r
dataset
```

```r
library(caTools)
```

```r
set.seed(123)
split = sample.split(dataset$Profit, SplitRatio = 0.8)
```

```r
split
```

```r
training_set = subset(dataset,split == TRUE)
test_set = subset(dataset,split == FALSE)
```

```r
print(test_set)
```

```r
print(training_set)
```

```r
regressor = lm(formula = Profit ~ .,
              data = training_set)
```

```r
summary(regressor)
```

```r
y_pred = predict(regressor,newdata = test_set)
```

```r
y_pred
```

```r
regressor1 = lm(formula = Profit ~ R.D.Spend,
              data = training_set)
```

```r
y_pred1 = predict(regressor1,newdata = test_set)
```

```r
library(ggplot2)
```

```r
ggplot() +
    geom_point(aes(x = test_set$R.D.Spend,y = y_pred1),color = 'red') +
    geom_point(aes(x = test_set$R.D.Spend,y = y_pred),color = 'green') +
    geom_point(aes(x = test_set$R.D.Spend,y = test_set$Profit),color = 'blue')
```
